```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../../MatLab/Outputs/test.csv"
df = pd.read_csv(PATH)

df.head()
```

```python
df.columns
```

```python
matplotlib.rcParams['figure.dpi'] = 120

sns.lineplot(data=df,
             x='ScreenX',
             y='ScreenY',
             hue='Track',
             alpha=0.9,
             palette='tab20')

plt.legend([],[], frameon=False)
plt.xlabel("Position (X)")
plt.ylabel("Position (Y)")
plt.title("TrackMate Tracks on WH04")

outpath = "../../Figures/test_data_tracks.png"
plt.savefig(outpath, bbox_inches='tight')
```
